```python
from retinex import Retinex
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = np.uint8(plt.imread("test_images/lena_sonet.png") * 255)
```

```python
retinex = Retinex()

# Single-scale retinex
image_retinex_SSR = retinex.apply(image, kernels=151, sigmas=151)

# Multi-scale retinex
image_retinex_MSR = retinex.apply(image, kernels=151, sigmas=(5, 80, 151))

# Drawing
plt.figure(figsize=(15, 20))
plt.subplot(1, 3, 1)
plt.axis('off')
plt.title("Original")
plt.imshow(image, cmap='gray')
plt.subplot(1, 3, 2)
plt.axis('off')
plt.title("Single-scale retinex")
plt.imshow(image_retinex_SSR, cmap='gray')
plt.subplot(1, 3, 3)
plt.axis('off')
plt.title("Multi-scale retinex")
plt.imshow(image_retinex_MSR, cmap='gray')
plt.show()
```
